```python
# Feature engineering
data.columns
```

```python
x = data[['Average Cost for two', 'Has Table booking', 'Has Online delivery', 'Price range']]
x
```

```python
y= data[['Aggregate rating']]
y
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
x['Has Table booking'] = labelencoder.fit_transform(x['Has Table booking'])
x
```

```python
x['Has Online delivery'] = labelencoder.fit_transform(x['Has Online delivery'])
x
```

```python
labelencoder.classes_
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test,predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predictions_lr = lr.predict(x_test)
predictions_lr
```

```python
modelresults(predictions_lr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C" : [0.1,0.5],"kernel" : ["linear","poly"],"degree" : [2]}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(x_train,y_train)
```

```python
predictions_gridsvr = gridsvr.predict(x_test)
predictions_gridsvr
```

```python
modelresults(predictions_gridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
predictions_tree = grid_tree.predict(x_test)
predictions_tree
```

```python
modelresults(predictions_tree)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth" : [5,15], "n_estimators" : [2,5,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(x_train,y_train)
```

```python
predictions_randmforest = gridrfr.predict(x_test)
```

```python
modelresults(predictions_randmforest)
```

```python
gridrfr.best_params_
```

```python
gridsvr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors":[3,5,7,9]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```

```python
knnpreds = knn_grid_search.predict(x_test)
```

```python
modelresults(knnpreds)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada_param_grid = {"n_estimators": [50,100,200],"learning_rate": [0.1,0.5,1]}
ada_grid_search = GridSearchCV(ada, ada_param_grid)
```

```python
ada_grid_search.fit(x_train,y_train)
```

```python
adapreds = ada_grid_search.predict(x_test)
```
